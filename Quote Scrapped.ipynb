```python
# Importing Libraries

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
# Executting url.

url = "http://quotes.toscrape.com/page/1/"
```

```python
# Uesing Requests libraries for url connection.

response = requests.get(url)
```

```python
response = response.content
```

```python
# Uesing BS libraries for url connection.

soup = BeautifulSoup(response,'html.parser')
```

```python
 div = soup.find('div')
```

```python
# Scrapping the the website to get Quotes.

contents = div.find_all('div',{'class':'quote'})

for content in contents:
    quote = content.find('span').text
    print(quote)
```

```python
for content in contents:
    author = content.find('small').text
    print(author)
```

```python
contents = div.find_all('div',{'class':'quote'})

Content = []

for content in contents:
    quote = content.find('span').text
    author = content.find('small').text
    tags = content.find('meta')
    Tags = tags.attrs['content']
    Content.append([quote,author,Tags])
    print(Content)
```

```python
Content = []

for i in range (1,5):
    url = f"http://quotes.toscrape.com/page/{i}"
    response = requests.get(url)
    response = response.content
    soup = BeautifulSoup(response,'html.parser')
    div = soup.find('div')
    contents = div.find_all('div',{'class':'quote'})
    for content in contents:
        quote = content.find('span').text
        author = content.find('small').text
        tags = content.find('meta')
        Tags = tags.attrs['content']
        Content.append([quote,author,Tags])
        print(Content)
```

```python
df = pd.DataFrame(Content,columns = ['quote','author','Tags'])
```
